```python
# Convert date and time columns to datetime
df['fecha'] = pd.to_datetime(df['fecha'], format='%d/%m/%Y')
df['start_time'] = pd.to_datetime(df['fecha'].astype(str) + ' ' + df['desde']).round('H')
df['end_time'] = pd.to_datetime(df['fecha'].astype(str) + ' ' + df['hasta']).round('H')
```

```python
from tqdm import tqdm


# Filter the DataFrame
df_c = df[df['linea'] == 'LineaC']
df_c['pax_total'] = pd.to_numeric(df_c['pax_total'], errors='coerce').astype(float)

# Initialize the results list
results = []

# Track progress with tqdm
estaciones = df_c['estacion'].unique()
fechas = df_c['start_time'].unique()

# Wrap the outer loop with tqdm
for station in tqdm(estaciones, desc="Processing Stations"):
    for date in tqdm(fechas, desc="Processing Dates", leave=False):
        total_pax = df_c[(df_c['estacion'] == station) & 
                         (df_c['start_time'] == date)]['pax_total'].sum()
        results.append({
            'station': station,
            'fecha_de_inicio': date,
            'total_pax': total_pax
        })

# Convert results to DataFrame if needed
results_df = pd.DataFrame(results)
```
